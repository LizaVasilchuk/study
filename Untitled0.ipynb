```
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_half = cv2.imread("Lab3_half100.png", cv2.IMREAD_GRAYSCALE)
img_chess = cv2.imread("Lab3_chess100.png", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img_half, cmap = 'gray');
plt.subplot(122)
#img_half
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show();
#img_chess
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img_chess, cmap = 'gray');
plt.subplot(122)
hist_chess = cv2.calcHist([img_chess], [0], None, [256], [0, 256])
plt.plot(hist_chess); plt.xlim([0, 255]); plt.grid(); plt.show();
plt.show();
```

```

half_blur3 = cv2.GaussianBlur(img_half,(3,3),0)
half_blur5 = cv2.GaussianBlur(img_half,(5,5),0)
half_blur9 = cv2.GaussianBlur(img_half,(9,9),0)
half_blur15 = cv2.GaussianBlur(img_half,(15,15),0)

chess_blur3 = cv2.GaussianBlur(img_chess,(3,3),0)
chess_blur5 = cv2.GaussianBlur(img_chess,(5,5),0)
chess_blur9 = cv2.GaussianBlur(img_chess,(9,9),0)
chess_blur15 =cv2.GaussianBlur(img_chess,(15,15),0)

half_blur3_hist = cv2.calcHist([half_blur3], [0], None, [256], [0, 256])
half_blur5_hist = cv2.calcHist([half_blur5], [0], None, [256], [0, 256])
half_blur9_hist = cv2.calcHist([half_blur9], [0], None, [256], [0, 256])
half_blur15_hist = cv2.calcHist([half_blur15], [0], None, [256], [0, 256])

chess_blur3_hist = cv2.calcHist([chess_blur3], [0], None, [256], [0, 256])
chess_blur5_hist = cv2.calcHist([chess_blur5], [0], None, [256], [0, 256])
chess_blur9_hist = cv2.calcHist([chess_blur9], [0], None, [256], [0, 256])
chess_blur15_hist = cv2.calcHist([chess_blur15], [0], None, [256], [0, 256])

plt.figure(figsize=(20,10))
plt.subplot(421),plt.imshow(half_blur3, cmap='gray'),plt.title('Gaussian Blur 3x3')
plt.xticks([]), plt.yticks([])
plt.subplot(422),plt.plot(half_blur3_hist), plt.title('Gistogram 3x3');
plt.subplot(423),plt.imshow(half_blur5, cmap='gray'),plt.title('Gaussian Blur 5x5')
plt.xticks([]), plt.yticks([])
plt.subplot(424),plt.plot(half_blur5_hist), plt.title('Gistogram 5x5');
plt.subplot(425),plt.imshow(half_blur9, cmap='gray'),plt.title('Gaussian Blur 9x9')
plt.xticks([]), plt.yticks([])
plt.subplot(426),plt.plot(half_blur5_hist), plt.title('Gistogram 9x9');
plt.subplot(427),plt.imshow(half_blur15, cmap='gray'),plt.title('Gaussian Blur 15x15')
plt.xticks([]), plt.yticks([])
plt.subplot(428),plt.plot(half_blur5_hist), plt.title('Gistogram 15x15');


plt.figure(figsize=(20,10))
plt.subplot(421),plt.imshow(chess_blur3, cmap='gray'),plt.title('Gaussian Blur 3x3')
plt.xticks([]), plt.yticks([])
plt.subplot(422),plt.plot(chess_blur3_hist), plt.title('Gistogram 3x3');
plt.subplot(423),plt.imshow(chess_blur5, cmap='gray'),plt.title('Gaussian Blur 5x5')
plt.xticks([]), plt.yticks([])
plt.subplot(424),plt.plot(chess_blur5_hist), plt.title('Gistogram 5x5');
plt.subplot(425),plt.imshow(chess_blur9, cmap='gray'),plt.title('Gaussian Blur 9x9')
plt.xticks([]), plt.yticks([])
plt.subplot(426),plt.plot(chess_blur9_hist), plt.title('Gistogram 9x9');
plt.subplot(427),plt.imshow(chess_blur15, cmap='gray'),plt.title('Gaussian Blur 15x15')
plt.xticks([]), plt.yticks([])
plt.subplot(428),plt.plot(chess_blur15_hist), plt.title('Gistogram 15x15');
plt.show()
```
