```
import cv2
import numpy as np
import matplotlib.pyplot as plt
print('Color BW')
img_bw = cv2.imread("1.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(7,7))
plt.imshow(img_bw, cmap='gray')
plt.show();

print('Color BGR')
img_bgr = cv2.imread("22.jpg")
#img_bgr_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))
plt.imshow(img_bgr)
plt.show();

print('Color RGB')
img_bgr = cv2.imread("22.jpg")
img_bgr_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))
plt.imshow(img_bgr_rgb)
plt.show();

print('Color HSV')
img_bgr = cv2.imread("22.jpg")
img_bgr_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(10, 10))
plt.imshow(img_bgr_hsv)
plt.show();

print('Color chanel for RGB')
plt.figure(figsize=(10, 10))

plt.subplot(131);
R = np.zeros(img_bgr_rgb.shape, dtype='uint8')
R[:, :, 0] = img_bgr_rgb[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_bgr_rgb.shape, dtype='uint8')
G[:, :, 1] = img_bgr_rgb[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_bgr_rgb.shape, dtype='uint8')
B[:, :, 2] = img_bgr_rgb[:, :, 2]
plt.imshow(B)

plt.show()

print('Color chanel for HSV')
plt.figure(figsize=(10, 10))

plt.subplot(131);
R = np.zeros(img_bgr_hsv.shape, dtype='uint8')
R[:, :, 0] = img_bgr_hsv[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_hsv.shape, dtype='uint8')
G[:, :, 1] = img_bgr_hsv[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_hsv.shape, dtype='uint8')
B[:, :, 2] = img_bgr_hsv[:, :, 2]
plt.imshow(B)

plt.show()

print('Area for BW')
area = img_bw.shape
a = area[0]/2
b = area[1]/2
img_area_bw = img_bw[0 : int(a), 0 : int(b)]
plt.imshow(img_area_bw, cmap = "gray")
plt.show()

cv2.imwrite("./img_area_bw.png", img_area_bw)
```
